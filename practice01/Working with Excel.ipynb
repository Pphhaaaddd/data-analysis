Getting all sheets separatly

```python
import numpy as np
import pandas as pd
xls = pd.ExcelFile('Jan18.xlsx')
df1 = xls.parse(0)
df2 = xls.parse(1)
df3 = xls.parse(2)

#First sheet (Uttam)
df1[:5]
```

```python
#Dropping unwanted rows/cols
#Changing First row as the keys
df1.columns = df1.iloc[0]
df1 = df1[1:]

#Removing unwanted column
df1.drop(df1.columns[[4]],inplace=True, axis=1)
df1[:5]
```

```python
df1['Date'] = df1.index
df1.index = range(len(df1))


#Rearranging Columns
cols = list(df1.columns.values)
cols = cols[-1:] + cols[:-1]
df1 = df1[cols]

#Renaming
df1=df1.rename(columns = {'DR':'Debit','CR':'Credit'})
df1.index.names = ['Index']
df1.columns.names = ['']
df1[0:9]
```

```python
#Replace NaN numbers values with 0
df1.iloc[:,[2,3]]=df1.fillna(0)

#Fill NaN Dates with previous date
df1.iloc[:,0]=df1.fillna(method='ffill')

#Change the numbers to float64 type
df1['Debit'] = df1['Debit'].astype('float64')
df1['Credit'] = df1['Credit'].astype('float64')
df1.iloc[:,[4]] = df1.iloc[:,[4]].astype('float64')

df1.iloc[:,[1]] = df1.iloc[:,[1]].astype('str')

#Checking maually entered balance
df1["Balance_new"] = df1['Debit'].cumsum() - df1['Credit'].cumsum() + df1.iloc[0, 4]

df1[:9]
```

```python
#Pivot to Date
df1_date = df1.pivot_table(['Credit', 'Debit'], ['Date'], aggfunc='sum')

cols = list(df1_date.columns.values)
cols = cols[-1:] + cols[:-1]
df1_date = df1_date[cols]

df1_date['Balance'] = df1_date['Debit'].cumsum() - df1_date['Credit'].cumsum() + df1.iloc[0, 4]
df1_date
```
