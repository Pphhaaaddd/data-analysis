```python
import numpy as np
import pandas as pd
xls = pd.ExcelFile('Jan18.xlsx')
df1 = xls.parse(0)
df2 = xls.parse(1)
df3 = xls.parse(2)

#First sheet (Uttam)
df1[:9]
```

```python
#Dropping unwanted rows/cols
#Changing First row as the keys
df1.columns = df1.iloc[0]
df1 = df1[1:]

#Removing unwanted column
df1.drop(df1.columns[[4]],inplace=True, axis=1)
df1[:9]
```

```python
df1['Date'] = df1.index
df1.index = range(len(df1))


#Rearranging Columns
cols = list(df1.columns.values)
cols = cols[-1:] + cols[:-1]
df1 = df1[cols]

#Renaming
df1=df1.rename(columns = {'DR':'Debit','CR':'Credit'})
df1.index.names = ['Index']
df1.columns.names = ['']
df1[0:9]
```

```python
     
#Replace NaN numbers values with 0
df1.iloc[:,[1,2]]=df1.fillna(0)


df1["Balance_new"] = 0
df1.iloc[0,5] = df1.iloc[0, 2] - df1.iloc[0,3]

for i in range(1, len(df1)):
    # Get the previous balance as well as current payments
    prev_balance = df1.iloc[i-1,5]
    reveived = df1.iloc[i, 2]
    paid = df1.iloc[i,3]

    df1.iloc[i,5] = prev_balance + reveived - paid
df1

df1[:9]
```

```python
#Fill NaN number values with 0
df1.iloc[:,[2,3]]=df1.fillna(')

#Fill NaN Dates with previous date
df1.iloc[:,0]=df1.fillna(method='ffill')

#Change Date type
df1['Date'] = df1['Date'].astype('datetime64[ns]')
df1[:9]
```
